```python
import numpy as np
import matplotlib.pyplot as plt


#load data file
outfile_this = 'OPC-response_hr.dat'


data = np.loadtxt(outfile_this)

#Get data from column
Dp = data[:, 0]       #Column 1 (Diameters)
log10_I3 = data[:, 3] #Column 4 (Succinic Acid)

#PLOT 

plt.figure(figsize=(8, 5), facecolor='white')

plt.plot(Dp, log10_I3, '.', color='blue', markersize=2, label='Succinic Acid (Raw)')

#way to smooth data
avgval = np.copy(log10_I3)
npts = 4

#Loop to calculate the mean of the neighboring points
for ii in range(npts, len(log10_I3) - npts):
    # Grab the window and calculate average
    window = log10_I3[ii - npts : ii + npts + 1]
    avgval[ii] = np.mean(window)


plt.plot(Dp, avgval, '-', color='red', linewidth=0.75, label='Smoothed')


plt.xscale('log')           
plt.xlim(200, 6000)         
plt.ylim(1, 5)              
plt.xlabel('Diameter (nm)')
plt.ylabel('Intensity')
plt.title('Mie Curve for Succinic Acid')
plt.grid()
plt.legend()
plt.show()
```
